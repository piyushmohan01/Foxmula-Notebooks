```python
# both-pandas-and-numpy-need-to-be-imported
# np.random.randn(x)-randamozies-nos-x-times

import pandas as pd
import numpy as np

df_g = pd.DataFrame({'key' : list('bbacccb'),
                 'val' : np.random.randn(7)})

df_g
```

```python
# creating-dummies
# prefix(optional)
# astype-fix-datatype

dfn = pd.get_dummies(df_g['key'], prefix = 'bool').astype(int)
dfn
```
